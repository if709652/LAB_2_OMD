```python
def f_leer_archivo(param_archivo):
     # Leer archivo de datos y guardarlo en un DataFrame
    df_data = pd.read_excel( 'reporte.xlsx', sheet_name='Hoja1')
    df_data = df_data.loc[df_data['Type']!='balance']
    
    # elegir solo renglones en los que la columna type == buy | type == sell
    df_data = df_data[(df_data['Type'] == 'buy') | (df_data['Type'] == 'sell') |
                        (df_data['Type'] == 's/l')
                        | (df_data['Type'] == 't/p')]
    
    # Convertir en minusculas el nombre de las columnas
    df_data.columns = [list(df_data.columns)[i].lower()
                       for i in range(0, len(df_data.columns))]
    

    #print(df_data)
    return df_data
```

```python
df_archivo = f_leer_archivo(param_archivo="reporte.xlsx")
df_archivo = df_archivo.reset_index()
df_archivo.head()
```

```python
from datetime import  datetime
from datetime import timedelta

def dif_tiempos(data):

    df = pd.DataFrame(data=df_tiempos)
    df['opentime'] = pd.to_datetime(df_tiempos['opentime'])
    df['closetime'] = pd.to_datetime(df_tiempos['closetime'])

    Time = df['closetime'] - df['opentime']
    Time = Time.dt.seconds
    Time = pd.DataFrame(Time)
    df_archivo['Time'] = Time
    return df_archivo
```

```python
# obtenemos los pips de todos los inttrumentos uilizados y anexamos al dataframe
pips = [f_pip_size(param_ins = df_archivo['symbol'][i]) for i in range(len(df_archivo))]
df_archivo['pips'] =pips
```

```python
#calculamos las diferencias entre entrada y salida de la operacion y multiplicamos por su valor equivalente
#deltasprice = [n for n in range(len(newd)) if n%2 == 0]
piptotales = []
conta = 0
for i in range(0,len(df_archivo)):
    if df_archivo['symbol'][i] == 'buy':
        piptotales.append((df_archivo['closeprice'][i] - df_archivo['openprice'][i])*pips[i])
    else:
        piptotales.append((df_archivo['openprice'][i] - df_archivo['closeprice'][i])*pips[i])
    conta = conta + 1
df_archivo['piptotales'] = piptotales
df_archivo['profit_t'] = df_archivo['profit'].cumsum() 
df_archivo
```
